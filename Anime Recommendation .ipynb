```python
import pandas as pd
import numpy as np

anime = pd.read_csv("Anime Data/anime.csv")
```

```python
filtered_df = anime[anime['name'] == 'Haikyuu!! Second Season']['anime_id']
```

```python
import re

def clean_name(name):
    
    return re.sub("[^a-zA-Z0-9 ]", "", name)
    
```

```python
anime["cleaned_name"] = anime["name"].apply(clean_name)
```

```python
anime.head(20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Count vectorization
vectoriser = TfidfVectorizer(ngram_range = (1,2)) # we have a vector/ matrix

tfidf = vectoriser.fit_transform(anime["cleaned_name"])
```

```python
#Finding the similar animes using cosine similarity

from sklearn.metrics.pairwise import cosine_similarity

def search(name):
    name = clean_name(name)
    query_vector = vectoriser.transform([name])
    similarity = cosine_similarity(query_vector, tfidf).flatten()
    
    indices = np.argsort(similarity)[-5:]
    
#     indices = np.argpartition(similarity, -5)[-5:] # finds the five most similar items to our search term
    
    results = anime.iloc[indices][::-1]
    
    return results
    
```

```python
search("Naruto: Shippuuden")
```

```python
# Ratings Data 

ratings = pd.read_csv("Anime Data/rating.csv")
```

```python
# finding the similar users 

# users who like haikyu
similar_users = ratings[(ratings["anime_id"] == 28891) & (ratings["rating"] > 8)]["user_id"].unique()
similar_users
```

```python
# other anime that similiar users like
# vahi users jinko meri jaise pasand hai and unko baki sab kya pasand hai 

similar_user_recommendations = ratings[(ratings["user_id"].isin(similar_users))& (ratings["rating"] > 8)]["anime_id"]
similar_user_recommendations
```

```python
similar_user_recommendations.value_counts() # this shows that Haikyuu(28891) have maximum count as I like haikyu
```

```python
similar_user_recommendations = similar_user_recommendations.value_counts()/len(similar_users)
```

```python
# we only need more the 20% of the anime for the recommendations 
similar_user_recommendations = similar_user_recommendations[similar_user_recommendations > 0.20]
```

```python
similar_user_recommendations
```

```python
# checking all the other users who have rated the movie and in our recommendation data frame
'''
The condition being checked is whether each value in the "anime_id" column of the "ratings"
DataFrame is present in the index of another DataFrame called "similar_user_recommendations".
'''

# this checks if any one who has rated the anime is in our similar_user_recommendations dataframe and also given 
# rating more than 8
all_users = ratings[ratings["anime_id"].isin(similar_user_recommendations.index) & (ratings['rating'] > 8)]
all_users
```

```python
# calculate the percentage of these users 
all_users_recommendations = all_users['anime_id'].value_counts() / len(all_users["user_id"].unique())
```

```python
all_users_recommendations
```

```python
# viewing both the similar and all user recommendations
rec_percentages = pd.concat([similar_user_recommendations, all_users_recommendations], axis=1)
rec_percentages.columns = ["similar", "all"]
rec_percentages
```

```python
rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]
rec_percentages = rec_percentages.sort_values("score", ascending=False)
rec_percentages
```

```python
rec_percentage = rec_percentages.merge(anime, left_index = True, right_on = "anime_id")["name"]
```

```python
def anime_recommendation(animeID):
    
    # find users similar to us 
    similar_users = ratings[(ratings["anime_id"] == animeID) & (ratings["rating"] > 8)]["user_id"].unique()
    similar_user_recs = ratings[(ratings["user_id"].isin(similar_users)) & (ratings["rating"] > 8)]["anime_id"]
    
    #percentatge of the user who are similar to us and like the movie
    similar_user_recs = similar_user_recs.value_counts() / len(similar_users)
    similar_user_recs = similar_user_recs[similar_user_recs > .20]
    
    # all of the users and there recommendations
    all_users = ratings[ratings["anime_id"].isin(similar_user_recs.index) & (ratings['rating'] > 8)]
    all_users_recs = all_users['anime_id'].value_counts() / len(all_users["user_id"].unique())
    
    # find the percentage
    rec_percentages = pd.concat([similar_user_recs, all_users_recs], axis=1)
    rec_percentages.columns = ["similar", "all"]
    
    rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]
    
    rec_percentages = rec_percentages.sort_values("score", ascending=False)
    return rec_percentages.head(10).merge(anime, left_index=True, right_on="anime_id")["cleaned_name"]
    
```

```python
import ipywidgets as widgets
from IPython.display import display

anime_name_input = widgets.Text(
    description='Anime name:',
    disabled=False
)
recommendation_list = widgets.Output()

def on_type(data):
    with recommendation_list:
        recommendation_list.clear_output()
        name = data["new"]
        if len(name) > 3:
            results = search(name)
            animeID = results.iloc[0]["anime_id"]
            display(anime_recommendation(animeID))

anime_name_input.observe(on_type, names='value')

display(anime_name_input, recommendation_list)
```
